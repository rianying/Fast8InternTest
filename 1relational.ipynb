```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('1.xlsx')
```

```python
#Is there a relationship between 'TurnOver' (company turnover) and 'StatusKaryawan' (employee status)? 
#How does the turnover vary for different employee statuses?

print('Null values inside "JumlahKaryawan"  :',df['JumlahKaryawan'].isna().sum())
print('Null values inside "BidangUsaha"     :',df['BidangUsaha'].isna().sum())
df.dropna(subset=['BidangUsaha'])
df['JumlahKaryawan'] = pd.to_numeric(df['JumlahKaryawan'], errors='coerce')
correlation = df.groupby('BidangUsaha')['JumlahKaryawan'].mean().sort_values()
correlation_top_bottom = pd.concat([correlation.head(10), correlation.tail(10)])
correlation_top_bottom.plot(kind='bar')
plt.xlabel('Business Sector')
plt.ylabel('Average Number of Employees')
plt.title('Business Sector and Employee Count (TOP AND BOTTOM 10)')
plt.show()
```
