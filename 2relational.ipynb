```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('1.xlsx')
```

```python
#How does the 'CaraAbsensi' (attendance method) vary across different 'Provinsi' (provinces)? 
#Are there any regional preferences or patterns in attendance tracking methods?

print('Null values inside "CaraAbsensi" : ',df['CaraAbsensi'].isna().sum())
print('Null values inside "Provinsi"    : ',df['Provinsi'].isna().sum())
df = df.dropna(subset=['CaraAbsensi', 'Provinsi'])
attendance_by_province = df.groupby(['Provinsi', 'CaraAbsensi']).size().unstack().fillna(0)
attendance_by_province.plot(kind='bar', stacked=True)
plt.xlabel('Provinces')
plt.ylabel('Count')
plt.title('Attendance Methods across Provinces')
plt.legend(loc='upper right')
plt.show()
```
