```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('1.xlsx')
```

```python
#Do companies that apply 'FasilitasBPJS' (BPJS facilities) also tend to provide 'ApakahCutiBersamaMemotongCutiTahunan' (deduct annual leave for collective holidays)? 
#Is there any relationship between these two HR practices?

print('Null values under "FasilitasBPJS"                       :', df['FasilitasBPJS'].isna().sum())
print('Null values under "ApakahCutiBersamaMemotongCutiTahunan":', df['ApakahCutiBersamaMemotongCutiTahunan'].isna().sum())

df['ApakahCutiBersamaMemotongCutiTahunan'] = df['ApakahCutiBersamaMemotongCutiTahunan'].map({'Ya': 'Deducted', 'Tidak': 'Not Deducted'})
contingency_table = pd.crosstab(df['FasilitasBPJS'], df['ApakahCutiBersamaMemotongCutiTahunan'])

contingency_table.plot(kind='bar', stacked=True)
plt.xlabel('FasilitasBPJS')
plt.ylabel('Count')
plt.title('Relationship between BPJS Facilities and Deduction of Annual Leave for Collective Holidays')
plt.legend(loc='upper right')
plt.xticks(rotation=0)
plt.show()
```
